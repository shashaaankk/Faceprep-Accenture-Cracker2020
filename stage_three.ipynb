# Operation Choices

You are required to implement the following function.

int OperationChoices(int c, int a, int b )

The function accepts 3 positive integers ‘a’ , ‘b’ and ‘c ‘ as its arguments. Implement the function to return.
( a+ b ) , if c=1
( a - b ) , if c=2
( a * b ) , if c=3
(a / b) , if c =4

Assumption: All operations will result in integer output.

Example:

Input

c:1
a:12
b:16

Output:
Since ‘c’=1, (12+16) is performed which is equal to 28, hence 28 is returned.

Sample Input

 c: 2
 a: 16
 b: 20

Sample Output

-4


```python
def operatorChoices(a,b,c):
    if c==1:
       return a+b
    elif c==2:
       return a-b
    elif c==3:
      return a*b
    elif c==4:
      return int(a/b)

      
c=int(input())
a=int(input())
b=int(input())
op=operatorChoices(a,b,c)
print(op)
```

# Intermediate palindrome numbers
 Write a program that it takes a lower limit and upper limit as inputs and print all the intermediate palindrome numbers.

Test Cases:

TestCase 1:

Input :

10 80

Expected Result:

11 22 33 44 55 66 77


Test Case 2:

Input:

100 200

Expected Result:

101 111 121 131 141 151 161 171 181 191


      

```python
def is_palindrome(i):
    forw=str(i)
    rev=forw[::-1]
    if forw==rev:
       return True
    return False
  
m=int(input())
n=int(input())
for i in range(m,n+1):
    if(is_palindrome(i)):
      print(i,end=" ")
```

# Delete at specific position
Write a program to create a dynamic array and delete an element from an array from the specified position. If deletion is not possible, print "Deletion not possible.".

Sample Input:

5

1 2 3 4 5

2

Sample Output:

1 3 4 5


```python
def deleteNum(arr,x):
    arr.remove(arr[x-1])
    for i in range(len(arr)):
        print(arr[i],end=" ")
    
n=int(input())
arr=list(map(int,input().split()))
x=int(input())
deleteNum(arr,x)
```

# Segregate 0's and 1's
Write a program to segregate all the 0’s in the left side and 1’s in the right side in the same array.

Sample Input:

5

0 1 0 1 0

Sample Output:

0 0 0 1 1



```python
def segregator(arr):
    zeros=[]
    ones=[]
    for num in arr:
        if num==0:
          zeros.append(num)
        elif num==1:
          ones.append(num)
    for z in range(len(zeros)):
        print(zeros[z],end="")
    for o in range(len(ones)):
        print(ones[o],end="")
    



n=int(input())
arr=list(map(int,input().split()))
segregator(arr)
```

# Maximum Exponent
You are given a function,

Int MaxExponents (int a , int b);

You have to find and return the number between ‘a’ and ‘b’ ( range inclusive on both ends) which has the maximum exponent of 2.


The algorithm to find the number with maximum exponent of 2 between the given range is

Loop between ‘a’ and ‘b’. Let the looping variable be ‘i’.
Find the exponent (power) of 2 for each ‘i’ and store the number with maximum exponent of 2 , let say ‘max’. Set ‘max’ to ‘i’ only if ‘i’ has more exponent of 2 than ‘max’.
Return ‘max’.

Assumption: a <b

Note: If two or more numbers in the range have the same exponents of 2 , return the small number.

Example

Input:

7

12

Output:

8

Explanation:

Exponents of 2 in:

7-0

8-3

9-0

10-1

11-0

12-2

Hence maximum exponent if two is of 8.



```python
def maximumexponent(n,m):
    res=0
    for i in range(m,n,-1):
        if(i & (i-1) == 0):
            res=i
            break
            
    return res
  
n=int(input())
m=int(input())
op=maximumexponent(n,m)
print(op)
```
